```python
import cv2, math, pandas as pd
import mediapipe as mp
from tqdm import tqdm


PROCESS_EVERY_N_FRAMES = 2
RESIZE_WIDTH = 640
DEBOUNCE_SECONDS = 0.18
CONTACT_SCALE = 1.2
MAX_HOUGH_RADIUS = 60


cap = cv2.VideoCapture('video.mp4')
orig_w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
orig_h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS) or 30.0
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT) or 0)

scale = RESIZE_WIDTH / orig_w
RESIZE_HEIGHT = int(orig_h * scale)


mp_pose = mp.solutions.pose
pose = mp_pose.Pose(min_detection_confidence=0.5, model_complexity=0)

events = []
prev_ball = None
prev_player_center = None
last_touch_frame = -999
frame_idx = 0


debounce_frames = int(DEBOUNCE_SECONDS * fps)

pbar = tqdm(total=total_frames, desc='Processing frames')

while True:
    ret, frame = cap.read()
    if not ret:
        break


    if frame_idx % PROCESS_EVERY_N_FRAMES != 0:
        frame_idx += 1
        pbar.update(1)
        continue


    small = cv2.resize(frame, (RESIZE_WIDTH, RESIZE_HEIGHT))
    gray = cv2.cvtColor(small, cv2.COLOR_BGR2GRAY)


    circles = cv2.HoughCircles(gray, cv2.HOUGH_GRADIENT, dp=1.2,
                               minDist=20, param1=50, param2=30,
                               minRadius=5, maxRadius=MAX_HOUGH_RADIUS)
    bx = by = bradius = None
    if circles is not None:
        c = circles[0][0]
        bx, by, bradius = int(c[0]), int(c[1]), int(c[2])

        bx_orig = int(bx / scale)
        by_orig = int(by / scale)
        bradius_orig = int(bradius / scale)
    else:
        bx_orig = by_orig = bradius_orig = None


    left_pt = right_pt = player_center = None
    rgb = cv2.cvtColor(small, cv2.COLOR_BGR2RGB)
    res = pose.process(rgb)
    if res.pose_landmarks:
        lm = res.pose_landmarks.landmark

        def to_orig(x, y):
            return (int(x * RESIZE_WIDTH / scale), int(y * RESIZE_HEIGHT / scale))
        try:
            l_ank = lm[mp_pose.PoseLandmark.LEFT_ANKLE]
            r_ank = lm[mp_pose.PoseLandmark.RIGHT_ANKLE]
            left_pt = to_orig(l_ank.x, l_ank.y)
            right_pt = to_orig(r_ank.x, r_ank.y)
            lhip = lm[mp_pose.PoseLandmark.LEFT_HIP]
            rhip = lm[mp_pose.PoseLandmark.RIGHT_HIP]

            player_center = (int(((lhip.x + rhip.x)/2) * RESIZE_WIDTH / scale),
                             int(((lhip.y + rhip.y)/2) * RESIZE_HEIGHT / scale))
        except Exception:
            left_pt = right_pt = player_center = None


    if prev_ball is not None and bx_orig is not None:
        dx = bx_orig - prev_ball[0]; dy = by_orig - prev_ball[1]
        ball_speed_px_s = math.hypot(dx, dy) * (fps / PROCESS_EVERY_N_FRAMES)
    else:
        ball_speed_px_s = 0.0


    if bx_orig is not None and left_pt and right_pt:
        dist_left = math.hypot(bx_orig - left_pt[0], by_orig - left_pt[1])
        dist_right = math.hypot(bx_orig - right_pt[0], by_orig - right_pt[1])
        contact_thresh = max(12, int(CONTACT_SCALE * bradius_orig)) if bradius_orig else 24
        if (dist_left < contact_thresh or dist_right < contact_thresh):
            if frame_idx - last_touch_frame > debounce_frames:
                leg = 'Left' if dist_left < dist_right else 'Right'

                vel_px_s = 0.0
                if prev_player_center is not None and player_center is not None:
                    pdx = player_center[0] - prev_player_center[0]
                    pdy = player_center[1] - prev_player_center[1]
                    vel_px_s = math.hypot(pdx, pdy) * (fps / PROCESS_EVERY_N_FRAMES)
                timestamp = frame_idx / fps
                events.append({
                    'frame': frame_idx,
                    'time_s': timestamp,
                    'leg': leg,
                    'ball_speed_px_s': ball_speed_px_s,
                    'player_vel_px_s': vel_px_s
                })
                last_touch_frame = frame_idx


    if bx_orig is not None:
        prev_ball = (bx_orig, by_orig)
    if player_center is not None:
        prev_player_center = player_center

    frame_idx += 1
    pbar.update(1)

pbar.close()
cap.release()

import pandas as pd
df = pd.DataFrame(events)
df.to_csv('touch_events_fast.csv', index=False)
print('Events saved to touch_events_fast.csv, total events:', len(df))
print(df.head())
```

```python
import cv2, math, pandas as pd
from collections import deque


VIDEO_IN = 'video.mp4'
CSV = 'touch_events_fast.csv'
OUT = 'annotated_output.mp4'
POPUP_DURATION_SEC = 0.9
FPS = None


df = pd.read_csv(CSV)
events_by_frame = {int(r.frame): r.to_dict() for _, r in df.iterrows()}

cap = cv2.VideoCapture(VIDEO_IN)
orig_w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
orig_h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
FPS = cap.get(cv2.CAP_PROP_FPS) or 30.0

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(OUT, fourcc, FPS, (orig_w, orig_h))


active_popups = deque()

frame_idx = 0
left_count = 0
right_count = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break


    if frame_idx in events_by_frame:
        e = events_by_frame[frame_idx]
        if e['leg'].lower().startswith('l'):
            left_count += 1
        else:
            right_count += 1
        popup_text = f"Touch: {e['leg']} | vel: {int(e['player_vel_px_s'])} px/s | ball: {int(e['ball_speed_px_s'])} px/s"
        popup = {'text': popup_text, 'end_frame': frame_idx + int(POPUP_DURATION_SEC * FPS)}
        active_popups.append(popup)


    cv2.putText(frame, f'Left: {left_count}', (12,40), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (50,220,50), 2)
    cv2.putText(frame, f'Right: {right_count}', (12,80), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (50,220,50), 2)


    while active_popups and active_popups[0]['end_frame'] < frame_idx:
        active_popups.popleft()

    y0 = 120
    for i, p in enumerate(list(active_popups)[-3:]):
        cv2.rectangle(frame, (10, y0 + i*40 - 28), (760, y0 + i*40 + 6), (0,0,0), -1)
        cv2.putText(frame, p['text'], (20, y0 + i*40), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255,255,255), 2)

    out.write(frame)
    frame_idx += 1

cap.release()
out.release()
print('Annotated video saved to', OUT)
```

```python
import cv2, numpy as np, pandas as pd, math
df = pd.read_csv('touch_events_fast.csv')
cap = cv2.VideoCapture('video.mp4')
FPS = cap.get(cv2.CAP_PROP_FPS) or 30.0

def get_frame_at(idx):
    cap.set(cv2.CAP_PROP_POS_FRAMES, idx)
    ret, f = cap.read()
    return f if ret else None

rotation_results = []
PATCH_RADIUS = 28

for _, row in df.iterrows():
    fidx = int(row.frame)

    frames = []
    for off in [-1, 0, 1]:
        fr = get_frame_at(max(0, fidx + off))
        if fr is not None:
            frames.append(fr)
    if len(frames) < 2:
        rotation_results.append({'frame': fidx, 'rotation': 'Unknown', 'score': 0.0})
        continue


    mid = frames[len(frames)//2]
    gray = cv2.cvtColor(mid, cv2.COLOR_BGR2GRAY)
    circles = cv2.HoughCircles(gray, cv2.HOUGH_GRADIENT, dp=1.2, minDist=20,
                               param1=50, param2=25, minRadius=5, maxRadius=80)
    if circles is None:
        rotation_results.append({'frame': fidx, 'rotation': 'Unknown', 'score': 0.0})
        continue
    x,y,r = map(int, circles[0][0][:3])


    patches = []
    for fr in frames:
        h,w = fr.shape[:2]
        x0, y0 = max(0, x-PATCH_RADIUS), max(0, y-PATCH_RADIUS)
        x1, y1 = min(w, x+PATCH_RADIUS), min(h, y+PATCH_RADIUS)
        p = fr[y0:y1, x0:x1]
        if p.shape[0] == 0 or p.shape[1] == 0:
            continue
        patches.append(cv2.cvtColor(p, cv2.COLOR_BGR2GRAY))

    if len(patches) < 2:
        rotation_results.append({'frame': fidx, 'rotation': 'Unknown', 'score': 0.0})
        continue


    flow = cv2.calcOpticalFlowFarneback(patches[0], patches[1], None,
                                        pyr_scale=0.5, levels=3, winsize=15,
                                        iterations=3, poly_n=5, poly_sigma=1.2, flags=0)
    h_p, w_p = flow.shape[:2]
    cx, cy = w_p//2, h_p//2
    tang_sum = 0.0; cnt = 0
    for i in range(0, h_p, 3):
        for j in range(0, w_p, 3):
            fx, fy = flow[i,j]
            dx = j - cx; dy = i - cy
            denom = math.hypot(dx, dy) + 1e-6
            tang = (-dy*fx + dx*fy)/denom
            tang_sum += tang; cnt += 1
    avg_tang = tang_sum / max(1, cnt)
    rotation = 'Forward' if avg_tang > 0 else 'Backward'
    score = abs(avg_tang)
    rotation_results.append({'frame': fidx, 'rotation': rotation, 'score': score})

cap.release()
rot_df = pd.DataFrame(rotation_results)
rot_df.to_csv('rotation_estimates.csv', index=False)
print(rot_df.head())
```
